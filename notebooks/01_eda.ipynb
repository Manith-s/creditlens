Profile the Home Credit application data: missingness, class balance (~8% default), and the strongest raw predictors (the external-source scores).


```python
import sys; sys.path.append('..')
from src.common.io import load_home_credit_application
import pandas as pd, numpy as np
df = load_home_credit_application()
df.shape, df['TARGET'].mean()
```

```python
# univariate signal: correlation with default
num = df.select_dtypes('number').drop(columns=['SK_ID_CURR'])
num.corrwith(df['TARGET']).abs().sort_values(ascending=False).head(12)
```
